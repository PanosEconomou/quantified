# Takeout data visualisation

This is a visualization of the takeout data denoting the activity on google.

```python
# Import modules
import numpy as np
import pandas as pd
import operator
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Read data
data = pd.read_csv('Activities.csv')

# Extract the dates
dates = np.array([datetime.strptime(date,"%Y-%m-%d %H:%M:%S %Z") for date in data['Activity Timestamp'].to_numpy()])

# Extract the frequency of product use 
labels,frequencies = np.unique(data['Product Name'].to_numpy(),return_counts=True)
```

```python
# Plot a little histogram of the data
# Create a figure
fig = plt.figure(figsize=(15,6),dpi=200)
ax = fig.add_subplot(111)

idx, frequencies = np.array(sorted(enumerate(frequencies), key=operator.itemgetter(1))).T
labels = labels[idx]

ax.bar(labels,frequencies,color='deepskyblue')
plt.xticks(rotation=60);
```
